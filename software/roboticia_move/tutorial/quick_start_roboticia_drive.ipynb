```python
import math as mt
p = []
for i in l:
    for j in i[1]:
        (q,a,d) = j
        (x,y) = (mt.cos(mt.radians(a))*d,mt.sin(mt.radians(a))*d)
        p.append((x,y))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = [x[0] for x in p]
y = [x[1] for x in p]

plt.scatter(x, y, s=3)
plt.show()
```
